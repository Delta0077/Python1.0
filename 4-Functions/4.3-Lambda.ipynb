## Lambda Functions in Python



```python
# Lambda functions are anonymous functions defined with the lambda keyword. They can take any number of arguments but can only have one expression. The syntax is as follows:
# lambda arguments: expression
```

```python
add = lambda a,b: a + b
print(add(5, 3))  # Output: 8
```

```python
addition = lambda a,b,c: a + b + c
print(addition(1,2,3))
```

```python
numbers = [2, 3, 4, 5]

list(map(lambda x: x**2,numbers))
```
