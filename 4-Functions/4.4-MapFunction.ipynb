## The map() Function in Python


```python
# The map() function applies a given function to each item of an iterable (like a list) and returns a map object (which is an iterator). The syntax is as follows:
# map(function, iterable, ...)
# The function is applied to each item of the iterable, and the result is returned as a map object. You can convert the map object to a list or other iterable type if needed.
def square(x):
    return x * x
```

```python
numbers = [1, 2, 3, 4, 5]
list(map(square, numbers))
```

```python
## Lambda function with map
numbers=[1,2,3,4,5]
list(map(lambda x:x*x, numbers))
```

```python
## Can we map multiple iterables?
numbers1 = [1, 2, 3,2 ]
numbers2 = [4, 5, 6]
list(map(lambda x,y: x+y, numbers1, numbers2))
```

```python
## map() to convert a list of strings to integers
# Use map() to convert a list of strings to integers
str_numbers = ['1', '2', '3', '4', '5']
list(map(int, str_numbers)) ## Converts each string in the list to an integer
```

```python
words = ['hello', 'world', 'python']
list(map(str.upper, words))
```

```python
def get_name(person):
    return person['name']

people= [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Charlie', 'age': 35}
]

list(map(get_name, people))
```
