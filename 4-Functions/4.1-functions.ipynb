# Introduction to Functions

## Defination:

A function is a block of code that performs a specefic task. Functions help in organizing code, reusing code and improving readablity.

```python
## Syntax
def function_name(parameters):
    """Docstring describing the function."""
    # Function body
    return result
```

```python
## Why use functions?
## Code Reusability**: Functions allow you to reuse code without having to rewrite it. You can call the same function multiple times with different arguments.
##Modularity**: Functions help break down complex problems into smaller, manageable pieces. This makes your code easier to read and maintain.
##Abstraction**: Functions allow you to hide the implementation details and expose only the necessary interface. This makes it easier to use and understand the code.
##Testing and Debugging**: Functions can be tested independently, making it easier to identify and fix bugs. You can write unit tests for individual functions to ensure they work as expected.
##Improved Readability**: Functions can improve the readability of your code by giving meaningful names to blocks of code. This makes it easier for others (or yourself in the future) to understand what the code is doing.
```

```python
number = int(input("Enter a number: "))
def even_or_odd(num):
    """ This function finds even or odd number."""
    if (number % 2) == 0:
        return "Even"
    else:
        return "Odd"
print(f"The number is {even_or_odd(number)}.")
```

```python
## Function with multiple parameters
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
def add_numbers(num1, num2):
    """ This function adds two numbers."""
    return num1 + num2
print(f"The sum of {num1} and {num2} is {add_numbers(num1, num2)}.")
```

```python
## Default parameter values
def greet(name="Guest"):
    print(f"Hello, {name}")
greet()  # Output: Hello, Guest
```

```python
### Variable length arguments
## Positional arguments

def print_numbers(*args):
    """ This function prints all the numbers passed as arguments."""
    for number in args:
        print(number)
print_numbers(1, 2, 3, 4, 5, "Hello, World!", [1, 2, 3], {"key": "value"})
```

```python
## Keyword arguments

def print_details(**kwargs):
    """ This function prints the details passed as keyword arguments."""
    for key,value in kwargs.items():
        print(f"{key}: {value}")
print_details(name="Alice", age=30, city="New York")
```

```python
def print_details(*args, **kwargs):
    for value in args:
        print(f"Positional argument:{value}")
    for key,value in kwargs.items():
        print(f"Keyword argument - {key}: {value}")
print_details(1,2,3,4,5,"Mr Deep", name="Alice", age=30, city="New York")
```

```python
### Return statement
def square(num):
    """ This function returns the square of a number."""
    return num ** 2
result = square(5)
print(f"The square of 5 is {result}.")
```
