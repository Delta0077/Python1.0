### Encapsulation and Abstraction

```python
## Encapsulation Getter and Setter Methods
## Public, Protected, Private variables or access modifier



class Person:
    def __init__(self,name,age):
        self.name=name      ##Public variables
        self.age=age        ## Public variables

def get_name(person):
    return person.name

person=Person("Mr Deep", 30)
get_name(person)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self.__name=name      ##Private variables
        self.__age=age        ##Private variables
        self.gender=gender    ## Public variables

def get_name(person):
    return person.__name

person=Person("Mr Deep", 32, "Male")
get_name(person)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self._name=name      ##Protected variables
        self._age=age        ##Protected variables
        self.gender=gender   ##Public variables

class Employee(Person):  ## Derived class of Person Class
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)

employee=Employee("Mr Deep",30,"Male")
print(employee._name)
```

```python
## Encapsulation with Getter and Setter Method
class Person:
    def __init__(self,name,age):
        self.__name=name ## Private access modifier or variable
        self.__age=age ## Private variable
    
    ## Getter method for name
    def get_name(self):
        return self.__name
    
    ## Setter method for name
    def set_name(self,name):
        self.__name=name
    
    ## Getter method for age
    def get_age(self):
        return self.__age
    
    ## Setter method for age
    def set_age(self,age):
        if age > 0:
            self.__age = age
        else:
            print("Age cant be negative.")
    
person=Person("Mr Deep", 30)

## Access and modify private variables using getter and setter

print(person.get_name())
print(person.get_age())

person.set_age(50)
print(person.get_age())

person.set_age(-2)
```
