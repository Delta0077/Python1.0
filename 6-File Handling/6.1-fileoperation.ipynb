```python
## Read a whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)  # Output: Contents of the file
```

```python
## Read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) # Output: Each line of the file without leading/trailing whitespace
```

```python
## Writting a file(Overwritting)

with open('example.txt', 'w') as file:
    file.write('This is a new line of text.\n')  # This will overwrite the existing content
    file.write('Another line of text with different content.\n')
```

```python
## Write a file without overwriting (Appending)

with open('example.txt', 'a') as file:
    file.write('This line will be appended to the existing content.\n')  # This will append to the existing content
    file.write('Appending another line of text.\n')
```

```python
### Writing list of lines to a file
lines = ['First line of text.\n', 'Second line of text.\n', 'Third line of text.\n']
with open('example.txt', 'w') as file:
    file.writelines(lines)  # This will overwrite the existing content with the list of lines
```

```python
## Binary file handling

#writing binary data to a file
data = b'\x00\x01\x02\x03\x04'  # Example binary data
with open('example.bin', 'wb') as file:
    file.write(data)  # This will write binary data to the file
```

```python
# Reading binary data from a file
with open('example.bin', 'rb') as file:
    binary_content = file.read()
    print(binary_content)  # Output: b'\x00\x01\x02\x03\x04'
```

```python
### Read the content from a source text file and write it to a new text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read() # Read the content of the source file

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content) # Write the content to the destination

```

```python
## Read a text file and count the number of lines, words, and characters
with open('example.txt', 'r') as file:
    content = file.read()
    lines = content.splitlines()  # Split the content into lines
    words = content.split()        # Split the content into words
    characters = len(content)      # Count the number of characters

    print(f'Number of lines: {len(lines)}')       # Output: Number of lines in the file
    print(f'Number of words: {len(words)}')       # Output: Number of words in the file
    print(f'Number of characters: {characters}')  # Output: Number of characters in the file
```

```python
## w+ mode allows you to read and write to a file. If the file does not exist, it will be created. If the file already exists, it will be overwritten.
```

```python
### Writing and then reading a file
with open('example.txt', 'w+') as file:
    file.write('Hello world\n')
    file.write('This is a new line \n')

    ## move the file pointer to the beginning of the file before reading
    file.seek(0)

    ## Read the content of the file
    content=file.read()
    print(content)

 
 
    
```
