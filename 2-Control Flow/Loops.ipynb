```python
## While loop
## The while loop is used to execute a block of code repeatedly until a certain condition is met.

count = 0

while count < 5:
    print(count)
    count += 1
```

```python
## Loop control statements
## Loop control statements are used to control the flow of a loop. They are used to skip

## the current iteration of a loop, to exit a loop, or to continue to the next iteration of a loop.

## break statement
## The break statement is used to exit a loop when a certain condition is met.
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## Continue statement
## The continue statement is used to skip the current iteration of a loop when a certain condition is met.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass statement
## The pass statement is used to indicate that a block of code is intentionally left empty. It
## is often used as a placeholder for code that will be added later.
for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loops
## A nested loop is a loop that is contained within another loop. The inner loop is executed
## for each iteration of the outer loop.
for i in range(3):
    for j in range(2):
        print(f"i={i} and j={j}")
```

```python
## Examples- Calculate the sum of the first n natural numbers using a for loop and a while loop.

## while loop

n = int(input("Enter n: "))
sum = 0
count = 1
while count <=n:
    sum += count
    count += 1
print("Sum of the first 10 natural numbers is:", sum)
```

```python
## Example- Prime numbers between 1 and n

n = int(input("Enter n: "))
for num in range(1, n+1):
    if num > 1:
        for i in range(2, num):
            if num%i == 0:
                break
        else:
            print(num)
```
