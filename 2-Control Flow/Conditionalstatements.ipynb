```python
## if statement
age = 20

if age >= 18:
    print("You are allowed to vote.")
```

```python
## else statement
## The else statement is used to execute a block of code when the condition in the if statement is false.
age = 16
if age >= 18:
    print("You are allowed to vote.")
else:
    print("You are not allowed to vote.")
```

```python
## elif statement
## The elif statement is used to check multiple conditions. It stands for "else if".

age = 20
if age < 13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
## Nested conditional statements

## You can place one or more if statements inside another if statement. This is called nesting.

## Number even or odd and positive or negative
num=int(input("Enter a number: "))

if num>=0:
    print("The number is positive.")
    if num%2==0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is negative.")
```

```python
## Practical example

## Determine if a year is a leap year using nested if statements.

year=int(input("Enter a year:"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year, "is a leap year.")
        else:
            print(year, "is not a leap year.")
    else:
        print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
## Assignment
## Simple Calculator using conditional statements
## Take user input for two numbers and an operator (+, -, *, /) and perform the corresponding operation.

num1=float(input("Enter first number:"))
num2=float(input("Enter second number:"))
operation = input("Enetr operation (+, -, *, /):")
if operation == "+":
    result = num1 + num2
    print("Result:", result)
elif operation == "-":
    result = num1 - num2
    print("Result:", result)
elif operation == "*":
    result = num1 * num2
    print("Result:", result)
elif operation == "/":
    if num2 !=0:
        result = num1 / num2
        
    else:
        print("Error: Division by zero is not allowed.")
else:
    result = "Invalid operator."

print("Result:", result)
```
