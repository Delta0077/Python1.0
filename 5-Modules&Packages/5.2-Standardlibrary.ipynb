## Standard Library Overview


```python
## Python's Standard Library is a vast collection of modules and packages that come bundled with Python,
## providing a wide range of functionalities out of the box. These modules cover various domains such as file handling, data manipulation, networking, and more. The Standard Library allows developers to perform common tasks without needing to install additional packages, making Python a powerful and versatile programming language. Some popular modules in the Standard Library include `os` for
## interacting with the operating system, `sys` for accessing system-specific parameters and functions, `math` for mathematical operations, and `datetime` for working with dates and times. By leveraging the Standard Library, developers can save time and effort while building robust applications.

import array
arr = array.array('i', [1, 2, 3, 4, 5])
print(arr)  # Output: array('i', [1, 2, 3, 4, 5])
```

```python
## random

import random
print(random.randint(1, 10))  # Output: A random integer between 1 and 10
print(random.choice(['apple', 'banana', 'cherry']))  # Output: A random choice from the list
```

```python
## Data Serialization
import json
data = {'name': 'Alice', 'age': 30, 'city': 'New York'}

json_str = json.dumps(data)  # Serialize to JSON string
print(json_str)  # Output: {"name": "Alice", "age": 30, "city": "New York"}
print(type(json_str))  # Output: <class 'str'>

parsed_data = json.loads(json_str)  # Deserialize back to Python object
print(parsed_data)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}
print(type(parsed_data))  # Output: <class 'dict'>
```

```python
## CSV

import csv

with open('example.csv', mode='w', newline='') as file:
    writer=csv.writer(file) ## Create a CSV writer object
    writer.writerow(['Name', 'age']) ## Write the header row
    writer.writerow(['Alice', 32]) ## Write a data row

with open('example.csv', mode='r') as file:
    reader=csv.reader(file) ## Create a CSV reader object
    for row in reader:
        print(row) ## Output: ['Name', 'age'] and ['Alice', '32']



```

```python
## datetime

from datetime import datetime,timedelta
now = datetime.now()
print(now)  # Output: Current date and time

yesterday = now -timedelta(days=1) 
print(yesterday)  # Output: Date and time of yesterday
```

```python
## time
import time
print(time.time())  # Output: Current time in seconds since the epoch
time.sleep(2)  # Pause execution for 2 seconds
print(time.time())  # Output: Current time in seconds since the epoch (after sleep)
```

```python
## Regular Expressions
import re

pattern = r'\d+'  # Pattern to match one or more digits
text = 'The price is 100 dollars'
match = re.search(pattern, text)
print(match.group())  # Output: '100'

```
