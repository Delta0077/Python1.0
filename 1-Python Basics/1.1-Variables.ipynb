```python
## Understanding variable types
## Python is a dynamically typed language, which means you don't need to declare the type of a variable. The type is inferred based on the value assigned to it.
```

```python
## Type checking and type conversion

age = 30
print(type(age))

# Type conversion
age_str = str(age)
print(type(age_str))
```

```python
age="30"
print(type(age))
```

```python
int(age)
print(type(int(age)))
```

```python
## Dynamic typing
## Python allows the type of a variable to change as the program executes. 
## This means you can assign a value of one type to a variable and later assign a value of a different type to the same variable.

var=10 #int
print(var, type(var))

var="Hello"
print(var, type(var))

var=5.7
print(var, type(var))
```

```python
## Simple calculator
num1 = float(input("Enter first number:"))
num2 = float(input("Enter second number:"))

sum = num1 + num2
diff = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum)
print("Difference:", diff)
print("Product:", product)
print("Quotient:", quotient)
```
