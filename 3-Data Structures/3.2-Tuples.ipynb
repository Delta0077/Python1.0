```python
# Tuples are immutable sequences, similar to lists but cannot be modified after creation
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple)
print(type(my_tuple))

# Accessing tuple elements (same as lists)
print(my_tuple[0])
print(my_tuple[-1])

# Tuple slicing
print(my_tuple[1:3])
```

```python
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
mixed_tuple = (1, "Hello", 3.14, True)
print(mixed_tuple)
```

```python
print(type(mixed_tuple))
```

```python
## Immutable nature of tuples
## Means you cannot change the elements of a tuple after it has been created or assign new values to existing elements. However, you can create a new tuple by concatenating or slicing existing tuples.

lst = [1, 2, 3]
lst[1] = "Mr Deep"
print(lst)
print(lst[1])

```

```python
numbers
numbers[1] = "Hello"
```

```python
## Tuple methods
# Tuples have only two built-in methods: count() and index()

print(numbers.count(10))
print(numbers.index(30))
```

```python
## Packing and unpacking tuples

## Packing
packed_tuple = 1, "Hello", 3.14, True
print(packed_tuple)
```

```python
## unpacking a tuple
a,b,c,d = packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
## Unpacking with * operator
numbers = (1,2,3,4,5,6)
first, middle, *last = numbers
print(first)
print(middle)
print(last)
```

```python

## Nested list
lst = [[1,2,3],[4,5,6],[7,8,9]]
print(lst[0][::])
```

```python
## Nested tuples
nested_tuple = ((1,2,3),("a","b","c"),(True,False))
print(nested_tuple[1][1])
```

```python
## Iterating through a tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end = " ")
    print()
```
