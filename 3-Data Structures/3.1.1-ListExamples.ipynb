# Example 1. Manage A To Do List

* Create a To Do List to Keep Track of Tasks

```python
to_do_list = ["Buy groceries", "Walk the dog", "Read a book"]

## Adding items to the list
to_do_list.append("Exercise")
to_do_list.append("Call a friend")

## Removing items from the list
to_do_list.remove("Walk the dog")


## Checking if an item is in the list
if "Read a book" in to_do_list:
    print("Don't forget to read a book!")

print("to_do_list remaining items:")
for item in to_do_list:
    print(f"- {item}")
```

# Example 2: Organizing Student Grades
* Create a list to store and calculate average grades for students

```python
# Organizing student grades using a dictionary
grades = [85, 90,78, 92, 88]

## Adding a new student grade
grades.append(95)

# Calculating the average grade
average_grade = sum(grades) / len(grades) # inbuilt function to calculate the sum of the list and len to calculate the number of items in the list
print(f"Average grade: {average_grade:.2f}")

# Finding the highest and lowest grades
highest_grade = max(grades) # inbuilt function to find the maximum value in the list
lowest_grade = min(grades) # inbuilt function to find the minimum value in the list
print(f"Highest grade: {highest_grade}")
print(f"Lowest grade: {lowest_grade}")
```

# Example 3: Managing an Inventory
* Use a list to manage inventory items in a store 

```python
# Managing an inventory
inventory = ["Apples", "Bananas", "Grapes"]

## Adding new items to the inventory
inventory.append("Mangoes")

# Removing an item from the inventory
inventory.remove("Bananas")

# Checking if an item is in the inventory
if "Oranges" in inventory:
    print("Oranges are in stock!")
else:
    print("Oranges are out of stock!")
print("Current inventory:")
for item in inventory:
    print(f"- {item}")

```

# Example 4: Collecting User Feedback
* Use a list to collect and analyze user feedback

```python
# Collecting user feedback
feedback = ["great service", "friendly staff", "clean environment"]

## Adding new feedback
feedback.append("will visit again")

# Counting specific feedback
positive_feedback_count = sum(1 for comment in feedback if "great" in comment.lower() or "friendly" in comment.lower()  or "clean" in comment.lower())
print(f"Number of positive feedback: {positive_feedback_count}")

# Printing all feedback comments
print("Customer feedback:")
for comment in feedback:
    print(f"- {comment}")
```
