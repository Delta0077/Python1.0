```python
## Dictionaries 
# A dictionary is a collection of key-value pairs. Each key is unique and maps to a value.
```

```python
## Creating a dictionary
empty_dict = {}
print(type(empty_dict))  # Output: <class 'dict'>
```

```python
## Another way to create a dictionary
empty_dict = dict()
empty_dict
```

```python
student={"name":"Aditya","age":30,"grade":"A"}
print(student)  # Output: {'name': 'Aditya', 'age': 30, 'grade': 'A'}
print(type(student))  # Output: <class 'dict'>
```

```python
## Error: Keys must be unique
student={"name": "Krish","age": 30, "name": "Aditya"}
print(student)  # Output: {'name': 'Aditya', 'age': 30} value of 'name' is overwritten by the second assignment i.e. 'Aditya'
```

```python
## accessing values in a dictionary
student={"name":"Aditya","age":30,"grade":"A"}
print(student)
```

```python
## Accessing values using keys
print(student["grade"])  # Output: A
print(student["age"])    # Output: 30
```

```python
## Accessing using get() method
print(student)
print(student.get("grade"))
print(student.get("last_name"))
print(student .get("last_name", "Not available"))  # Output: Not Found
```

```python
## Modifying values in a dictionary
## Dictionaries are mutable, which means you can change their contents after they have been created.

student["age"] = 33 ## Modifying the value of the existing key 'age'
print(student)  # Output: {'name': 'Aditya', 'age': 33, 'grade': 'A'}
student["address"] = "123 Main St Cityville USA 12345" ## Adding a new key-value pair
print(student)  # Output: {'name': 'Aditya', 'age': 33, 'grade': 'A', 'address': '123 Main St Cityville USA 12345'}
```

```python
## deleting a key-value pair
del student["grade"]

print(student)  # Output: {'name': 'Aditya', 'age': 33, 'address': '123 Main St Cityville USA 12345'}
```

```python
## Dictionary methods

keys = student.keys()
print(keys)  # Output: dict_keys(['name', 'age', 'address'])
values = student.values()
print(values)  # Output: dict_values(['Aditya', 33, '123 Main St Cityville USA 12345'])

items = student.items()
print(items)  # Output: dict_items([('name', 'Aditya'), ('age', 33), ('address', '123 Main St Cityville USA 12345')])

```

```python
## Shalow copy of a dictionary

student_copy = student
print(student)
print(student_copy)
```

```python
student_copy["name"] = "Krish"
print(student)  # Output: {'name': 'Krish', 'age': 33, 'address': '123 Main St Cityville USA 12345'}
print(student_copy)  # Output: {'name': 'Krish', 'age': 33, 'address': '123 Main St Cityville USA 12345'}
```

```python
student_copy1 = student.copy()  # Creates a shallow copy of the dictionary
print(student_copy1)  # Output: {'name': 'Krish', 'age': 33, 'address': '123 Main St Cityville USA 12345'}
print(student)
```

```python
student["name"] = "Mr Deep"
print(student)  # Output: {'name': 'Mr Deep', 'age': 33, 'address': '123 Main St Cityville USA 12345'}
print(student_copy1)  # Output: {'name': 'Krish', 'age': 33, 'address': '123 Main St Cityville USA 12345'}
```

```python
### Iterating over a dictionary

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterating over values
for values in student.values():
    print(values)
```

```python
## Iterating over key-value pairs
for key,vaule in student.items():
    print(f"{key}: {vaule}")
```

```python
## Nested dictionaries
students = {
    "student1": {"name": "Aditya", "age": 33, "grade": "A"},
    "student2": {"name": "Krish", "age": 30, "grade": "B"}

}

print(students)
```

```python
## Accessing nested dictionary values
print(students["student1"]["name"])
print(students["student1"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionary key-value pairs
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary comprehensions
squares = {x: x**2 for x in range(1,5)}
print(squares)
```

```python
## Conditional dictionary comprehensions
squares_even = {x: x**2 for x in range (1,10) if x % 2 == 0}
print(squares_even)
```

```python
## Practical example

## Use a dictionary to count the frequency of words in a text
text = "hello world hello"
word_count = {}
for word in text.split():
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
print(word_count)  # Output: {'hello': 2, 'world': 1}


## Use a dictionary to count the frequency of elements in a list
numbers = [1, 2, 2, 3, 3, 3]
number_count = {}

for number in numbers:
    if number in number_count:
        number_count[number] += 1
    else:
        number_count[number] = 1
    print(number_count)  # Output: {1: 1, 2: 2, 3: 3} ## The output is printed in each iteration to show the counting process
```

```python
ants = ["red", "black", "red", "red", "black"]

ant_count = {}

for ant in ants:
    if ant in ant_count:
        ant_count[ant] +=1
    else:
        ant_count[ant] = 1
    print(ant_count)  # Output: {'red': 3, 'black': 2}
```

```python
## Merging two dictionaries

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}
merged_dict = {**dict1, **dict2}
print(merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}
```
